```python
# @title FLUX.1 Setup
!pip install -U huggingface_hub requests

import os
import shutil
from huggingface_hub import login, hf_hub_download
import requests
from google.colab import userdata

# Get the Hugging Face login token from userdata
HF_LOGIN = userdata.get('HF_LOGIN')

if not HF_LOGIN:
    raise ValueError("HF_LOGIN not found in userdata. Please set it up in Google Colab.")

# Login to Hugging Face
login(HF_LOGIN)

# Define the paths
WORKSPACE = "/content/drive/MyDrive/ComfyUI" if 'USE_GOOGLE_DRIVE' in globals() and USE_GOOGLE_DRIVE else os.path.join(os.getcwd(), "ComfyUI")
UNET_DIR = os.path.join(WORKSPACE, "models", "unet")
CLIP_DIR = os.path.join(WORKSPACE, "models", "clip")
VAE_DIR = os.path.join(WORKSPACE, "models", "vae")

# Create directories if they don't exist
for dir_path in [UNET_DIR, CLIP_DIR, VAE_DIR]:
    os.makedirs(dir_path, exist_ok=True)

# Download FLUX.1 model
try:
    flux_filename = "flux1-dev.safetensors"
    dest_flux_path=os.path.join(UNET_DIR, flux_filename)
    if not os.path.exists(dest_flux_path):
      src_flux_path = hf_hub_download(repo_id="black-forest-labs/FLUX.1-dev", filename=flux_filename)
      shutil.copy2(src_flux_path, dest_flux_path)
      os.remove(src_flux_path)
      print(f"FLUX.1 model downloaded to: {dest_flux_path}")
    else:
      print(f"{dest_flux_path} is existed. skip the download.")
except Exception as e:
    print(f"Error downloading FLUX.1 model: {str(e)}")
    print("Please check your Hugging Face token and ensure you have access to the repository.")

# Download CLIP models
clip_urls = [
    "https://huggingface.co/comfyanonymous/flux_text_encoders/resolve/main/clip_l.safetensors",
    "https://huggingface.co/comfyanonymous/flux_text_encoders/resolve/main/t5xxl_fp16.safetensors",
    "https://huggingface.co/comfyanonymous/flux_text_encoders/resolve/main/t5xxl_fp8_e4m3fn.safetensors"
]

for url in clip_urls:
    clip_filename = url.split("/")[-1]
    try:
        dest_clip_path=os.path.join(CLIP_DIR, clip_filename)
        if not os.path.exists(dest_clip_path):
          src_clip_path = hf_hub_download(repo_id="comfyanonymous/flux_text_encoders", filename=clip_filename)
          shutil.copy2(src_clip_path, dest_clip_path)
          os.remove(src_clip_path)
          print(f"Downloaded {clip_filename} to {CLIP_DIR}")
        else:
          print(f"{dest_clip_path} is existed. skip the download.")
    except Exception as e:
        print(f"Error downloading {clip_filename}: {str(e)}")

# Download VAE
try:
    vae_filename = "ae.safetensors"
    dest_vae_path=os.path.join(VAE_DIR, vae_filename)
    if not os.path.exists(dest_clip_path):
      src_vae_path = hf_hub_download(repo_id="black-forest-labs/FLUX.1-schnell", filename=vae_filename)
      shutil.copy2(src_vae_path, dest_vae_path)
      os.remove(src_vae_path)
      print(f"Downloaded ae.safetensors to {VAE_DIR}")
    else:
      print(f"{dest_vae_path} is existed. skip the download.")
except Exception as e:
    print(f"Error downloading VAE: {str(e)}")

# Download Workflow image
workflow_url = "https://github.com/comfyanonymous/ComfyUI_examples/raw/master/flux/flux_dev_example.png"
try:
    response = requests.get(workflow_url)
    response.raise_for_status()
    with open(os.path.join(WORKSPACE, "flux_dev_example.png"), "wb") as f:
        f.write(response.content)
    print(f"Downloaded flux_dev_example.png to {WORKSPACE}")
except requests.exceptions.RequestException as e:
    print(f"Error downloading workflow image: {str(e)}")

print("Setup completed. Please check the output for any errors.")
```

### Run ComfyUI with cloudflared (Recommended Way)




```python
!wget -P ~ https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64.deb
!dpkg -i ~/cloudflared-linux-amd64.deb

import subprocess
import threading
import time
import socket
import urllib.request

def iframe_thread(port):
  while True:
      time.sleep(0.5)
      sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
      result = sock.connect_ex(('127.0.0.1', port))
      if result == 0:
        break
      sock.close()
  print("\nComfyUI finished loading, trying to launch cloudflared (if it gets stuck here cloudflared is having issues)\n")

  p = subprocess.Popen(["cloudflared", "tunnel", "--url", "http://127.0.0.1:{}".format(port)], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
  for line in p.stderr:
    l = line.decode()
    if "trycloudflare.com " in l:
      print("This is the URL to access ComfyUI:", l[l.find("http"):], end='')
    #print(l, end='')


threading.Thread(target=iframe_thread, daemon=True, args=(8188,)).start()

!python main.py --dont-print-server
```

### Run ComfyUI with colab iframe (use only in case the previous way with localtunnel doesn't work)

You should see the ui appear in an iframe. If you get a 403 error, it's your firefox settings or an extension that's messing things up.

If you want to open it in another window use the link.

Note that some UI features like live image previews won't work because the colab iframe blocks websockets.

```python
import threading
import time
import socket
def iframe_thread(port):
  while True:
      time.sleep(0.5)
      sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
      result = sock.connect_ex(('127.0.0.1', port))
      if result == 0:
        break
      sock.close()
  from google.colab import output
  output.serve_kernel_port_as_iframe(port, height=1024)
  print("to open it in a window you can open this link here:")
  output.serve_kernel_port_as_window(port)

threading.Thread(target=iframe_thread, daemon=True, args=(8188,)).start()

!python main.py --dont-print-server
```
